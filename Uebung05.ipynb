```python
def analyze_text(s: str = text) -> dict:
    words: list = s.split()
    joined_words: str = " ".join(words).lower()

    punc_marks: str = ".,!?;:"
    cleaned_text: str = ""

    for i in range(len(joined_words)):
        current_char: str = joined_words[i]

        if i > 0 and current_char in punc_marks and current_char == joined_words[i-1]:
            continue

        cleaned_text += current_char

    final_text: str = ""
    new_sentence: bool = True

    for i in cleaned_text:
        if new_sentence and i.isalpha():
            final_text += i.upper()
            new_sentence: bool = False
        else:
            final_text += i

            if i in ["!", ".", "?"]:
                new_sentence: bool = True

    return {
        "total_words": len(cleaned_text.split()),
        "total_unique_words": len(set(cleaned_text.split())),
        "longest_word": max(cleaned_text.split(), key=len),
        "shortes_word": min(cleaned_text.split(), key=len),
        "cleaned_final_text": final_text
    }

    
    
```

```python
analyze_text()
```

```python
def stats_from_tuples(d: list = daten) -> dict:
    ages: list = [age[1] for age in d]
    income: list = [income[2] for income in d]

    avg_age: float = sum(ages) / len(ages)
    avg_income: float = sum(income) / len(income)

    oldest: float = max(d, key=lambda x: x[1])
    youngest: float = min(d, key=lambda x: x[1])

    sorted_names: list = sorted([name[0] for name in d])
    over_avg: list = [people for people in d if people[2] > avg_income]

    return {
        "avg_age": avg_age,
        "avg_income": avg_income,
        "oldest": oldest,
        "youngest": youngest,
        "sorted_names": sorted_names,
        "over_avg": over_avg,
    }
    
```
